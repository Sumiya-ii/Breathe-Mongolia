```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.patches import Patch
from matplotlib.lines import Line2D
import calplot
```

```python
#loading datasets
mnb = pd.read_csv("/Users/bayarmaaorsoo/Desktop/Breathe Mongolia/Data Analysis/Ulaanbaatar Data Analysis/Data CSV/m.n.b.,-ulaanbaatar-air-quality.csv")
```

```python
#Converting data type of date (object to datetime)
mnb['date'] = pd.to_datetime(mnb['date'])
```

```python
#Doing same data cleaning for other data sets

#MNB
mnb.columns = mnb.columns.str.replace(' ','')
mnb = mnb.sort_values(by = 'date', ascending = False)
cols_to_convert_mnb = mnb.columns.drop('date')
mnb[cols_to_convert_mnb] = mnb[cols_to_convert_mnb].apply(pd.to_numeric, errors='coerce')
```

```python
# MNB PM2.5

# Main line plot
plt.figure(figsize=(25,10))
sns.lineplot(x = 'date', y = 'pm25', data =mnb, color = 'black', linewidth = 0.5)

# Some important date and PM2.5 values
plt.axvline(pd.Timestamp('2019-05-15'), linewidth = 2, color = 'b')
plt.text(pd.Timestamp('2019-05-25'), 420, "Ban of raw coal", horizontalalignment='left', fontsize=22, color='b')

#plt.axhline(y=57, color='deepskyblue', linestyle='dotted', linewidth = 4)
#plt.text(pd.Timestamp('2022-01-15'), 62, "WHO PM2.5 standard", horizontalalignment='left', fontsize=22, color='deepskyblue')

#plt.axhline(y=137, color='r', linestyle='dotted', linewidth = 4)
#plt.text(pd.Timestamp('2022-01-15'), 142, "Mongolian National PM2.5 standard", horizontalalignment='left', fontsize=22, color='red')

#Legends
legend_elements = [Patch(facecolor='darkred', label='Hazardous', alpha = 0.5),
                  Patch(facecolor='purple', label='Very unhealthy', alpha = 0.5),
                  Patch(facecolor='red', label='Unhealthy', alpha = 0.5),
                  Patch(facecolor='orange', label='Unhealthy for sensitive people', alpha = 0.5),
                  Patch(facecolor='yellow', label='Moderate', alpha = 0.5),
                  Patch(facecolor='green', label='Good', alpha = 0.5)]

plt.legend(handles=legend_elements, loc='upper right', fontsize = 14)

#AQI category
plt.axhspan(0, 50, facecolor='green', alpha=0.4)
plt.axhspan(50, 100, facecolor='yellow', alpha=0.4)
plt.axhspan(100, 150, facecolor='orange', alpha=0.4)
plt.axhspan(150, 200, facecolor='red', alpha=0.4)
plt.axhspan(200, 300, facecolor='purple', alpha=0.4)
plt.axhspan(300, 500, facecolor='darkred', alpha=0.4)

#Label

plt.xlabel("Date", fontsize = 15)
plt.ylabel("PM2.5", fontsize = 15)
plt.title("PM2.5 - Mongolian National Broadcaster", fontsize = 22)

# Yearly change

plt.text(pd.Timestamp('2015-04-30'), 370, "-12.7%", horizontalalignment='left', fontsize=22, color='red')
plt.text(pd.Timestamp('2016-04-1'), 400, "+20.9%", horizontalalignment='left', fontsize=22, color='green')
plt.text(pd.Timestamp('2017-05-10'), 400, "-20%", horizontalalignment='left', fontsize=22, color='red')
plt.text(pd.Timestamp('2018-04-30'), 370, "+4.4%", horizontalalignment='left', fontsize=22, color='green')
plt.text(pd.Timestamp('2019-05-20'), 330, "-16.8%", horizontalalignment='left', fontsize=22, color='red')
plt.text(pd.Timestamp('2020-04-30'), 300, "-3.3%", horizontalalignment='left', fontsize=22, color='red')
plt.text(pd.Timestamp('2021-05-10'), 280, "-4.8%", horizontalalignment='left', fontsize=22, color='red')
```
